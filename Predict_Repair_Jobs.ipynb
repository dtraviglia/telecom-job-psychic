#### Project Overview
Telecom services, such as voice, data, and video services, are delivered on various telecommunication networks to subscribers’ homes and businesses.  Failures and operating problems associated with the network and network equipment cause service problems in which some of them can be fixed remotely and others require dispatch of technicians to the site to repair and fix the problem(s).   Managing the available resources and their capacity in executing the repair jobs is quite challenging because the daily volume of dispatch needs (jobs demand) is largely unpredictable due to several factors such as the characteristics of the network and network equipment, nature of repairs needed, regulations, weather, seasonality, and characteristics of the labor force (such as union membership, skills, culture); among others.   On the other hand, if extra resources and capacity is secured, the cost to secure these resources in addition to unplanned overtime cost can be substantial.  If a lack of resources and capacity exists, then broken promises and negative customer experiences occur which typically affects customer satisfaction.  The ability to predict the jobs demand in advance for immediate and near short terms can provide substantial benefits in optimizing capacity, reducing network downtime and improving customer satisfaction.

## Objective

Develop a statistical model to predict the daily and 21-day period demands for telecom repair services to help in optimizing repairs resources capacity, improve operational effectiveness, reduce overtime cost and improve customer services. 

Our inital hypothesis is that the number of repair jobs is related to weather conditions plus seasonality or some other time related phenomena.  This project will help us identify and develop a model for repair jobs based based on this idea.

## Feature Selection Process

The EDA analysis revealed several interesting observations.  The most important ones were related to specific time patterns in the repairs jobs time series, stronger linear correlation of specific weather variables, and significance of holidays as related to the repairs jobs.  

**Conclusions:** 

Our findings included observing a repeated pattern in the number of repairs jobs within the week preceding any specific day.  This pattern, where the jobs counts are higher during the first three days of the week (peak), taper down toward the weekend, and become lower within the weekend (valley) led to the creation of more features that we are calling 'jobs-days-lags' as input variables for modeling.  Further, there were correlations not only for the weekdays versus weekends but also within the weekdays themselves so dummy variables for each day of the week were selected as input for the model development. 

Analysis of the holiday’s impact on the jobs also indicated a pattern of reduction during the day of the holiday and because we've seen a normal pattern of previous days jobs affecting the next days jobs, we're including an indicator for the day after the holiday. Binary variables for the holiday and the day after the holiday were created and selected for input as well.   Finally, extensive correlation analysis of the weather variables showed that one of the categorical variables that include weather conditions is strongly correlated with the repairs jobs.  This variable was also coded into dummy variables where the set is selected as input for consideration in the model development.

## Linear Regression - Building Model

The prepared data set was divided into a training and validation (or testing) data set.   The data from 01/01/2013 to 09/27/2014 were used to develop and train the model while the data from 09/28/2014 to 10/18/2014 were used to test the model.

** Conclusions: **

Two modeling approaches were attempted: Bayesian statistics and linear regression.  The bayesian statistics did not yield any conclusive results, possibly because the data is not from a normal distribution so our efforts were focused on linear regression.  Linear regression was selected because it is a better fit for the prediction problem at hand, provides opportunity to explain input drivers of the prediction to business (which is a business requirement) compared to a black box approach like neural networks, and provides manageable production implication for the model (see the note the model production deployment).  Finally, the ability to use the modeling algorithm itself as a vehicle to better select the input variables was a critical consideration since the reducing the number of input variables for the model,  as much as practically possible, is an important business requirement for deployment and productionalization. 

The initial set of the input variables selected during the data exploration and features selection stages were employed in the linear regressions.  Model optimization was performed throughout multiple steps of backward elimination where input variables are evaluated in the regression(s) and the ones that carry the least statistical significance were removed.  The first step included the whole set of the initial input variables selected during the data exploration and features selection then a subsequent regression model step is exercised to remove the weakest variable from the initial set.  A second step includes the whole initial set of variables minus the removed variable in the first step and likewise regression is used to remove the weakest input variables from the remaining set and so on.  A total of ten steps were used to reach the final set of input variables for the final regression model.  All variables elimination from the regressions was based on testing the hypothesis of no difference for the corresponding regression coefficients at the alpha level of 0.05.

## Measure Model Performance

The finalized regression model is tested using the data from 9/28/2014 to 10/18/2014 to assess the model performance, establish the control chart, and structure the scoring process.  The latter was divided into two parts: (1) scoring for the first 11 days, and (2) scoring for day 12 to day 21.  The reason for the separate scoring is the availability of forecasted weather data for the model input.  While the model prediction period is 21 days, only the first 11 days have forecasted weather data.  As a result, an imputation was developed to supplement the forecasted weather data for the second period prediction.  The imputation is specific to the rain variable because the final input variables included only rain from the set of weather data used during the model development process.  Since rain is a binary variable (1,0), the expected value of the rain variable during the first 11-day period is used to impute the missing forecasted rain variable needed for the model in the days within the second 10-day period.  

The model prediction was tested using the mean absolute pediction error (MAPE), which is the ratio of the mean absolute residuals during the 21-day predicted repairs jobs over the mean actual jobs during the same 21-day period.  The business requirement for the model prediction accuracy requires a performance criterion of MAPE of less than or equal to 0.4.  The linear regression developed in this project meets and exceeds the required MAPE level. 

```python
#create the test data set
testY = test["corejobs"].values
testX = test[trainX.columns].copy()
```

```python
#predict the corejobs for 21 days start from first day in the test dataset
predDt = testX.index.min()

#impute the weightage for the rainy days (rain is the only weather variable chosen from backward elimination processs)
#find number of rainy days from the 11 days weather forecast data
rainydays = testX.ix[predDt:predDt + np.timedelta64(11, 'D')]['wx_rain'].sum()

#get the co-ef for wx_rain variable from the model
wx_rainCoef = allvarlinregr.params.get('wx_rain')

#calculate the weightage 
wt_wx_rainCoef = (wx_rainCoef / 11) * rainydays

#print the calculated values
#print wx_rainCoef, wt_wx_rainCoef
```

```python
#initialize the list to store the predicted corejobs for 21 days
predCorejobs = []

#predict the corejobs for the first day
predCorejobs.append(allvarlinregr.predict(testX.ix[predDt]))

#loop the prediction for 21 days and append to the test dataset
for i in np.arange(1, 21):
    
    #convert the no of days to day offset
    j = np.timedelta64(i, 'D')
    
    #append the predicted corejobs & its lags to the test dataframe
    testX.loc[predDt + j, 'corejobs1'] = predCorejobs[i-1]
    
    if i >= 3:
        testX.loc[predDt + j, 'corejobs3'] = predCorejobs[i-3]
    if i >= 6:
        testX.loc[predDt + j, 'corejobs6'] = predCorejobs[i-6]   
    if i >= 11:
        #update the wx_rain based on 11 days forecast weather 
        if rainydays > 0.:
          testX.loc[predDt + j, 'wx_rain'] = 1
        else:
          testX.loc[predDt + j, 'wx_rain'] = 0

    #predict the corejobs
    yhat = allvarlinregr.predict(testX.ix[predDt + j])
    
    #add the imputed weightage for the wx_rain for raindays days
    if i >= 11 and rainydays > 0:
        predCorejobs.append(yhat - wx_rainCoef + wt_wx_rainCoef)  
    else:
        predCorejobs.append(yhat)
        
testX['pcorejobs'] = predCorejobs
#print predCorejobs        
```

```python
#calculate the residuals for test dataset
tresidual = testY - predCorejobs

#take absolute residuals
tabs_residual = np.abs(tresidual);

tmean_abs_residual = np.mean(tabs_residual)
tmin_abs_residual = np.min(tabs_residual)
tmax_abs_residual = np.max(tabs_residual)
tstd_abs_residual = np.std(tabs_residual)
tcorejob_actual = np.mean(testY)

terror_rate = tmean_abs_residual / tcorejob_actual
taccuracy = 1 - tmean_abs_residual / tcorejob_actual

#print residual.shape
#print 'residual', tresidual
#print 'abs residual', tabs_residual
#print 'mean residual', tmean_abs_residual
#print 'min residual', tmin_abs_residual
#print 'max residual', tmax_abs_residual
#print 'std residual', tstd_abs_residual
#print 'actual corejobs', tcorejob_actual

print 'error_rate', terror_rate
print 'accuracy', taccuracy
```

## Model Predictions and Control
Model utilization to predict the repairs jobs for a specific geographical area involves the two portion of the model scoring to generate the 21-day prediction on a daily basis.  So every day there is prediction for the 21 days in future (21 prediction values).  A control on the predicted daily repairs jobs is implemented to establish an upper and lower limits for the prediction in which values outside these limits are considered outliers (overshoot or undershoot).  When a predicted demand is an overshoot or undershoot the value is controlled by this threshold.  The current criteria for the limits are to set such that the lower limit equals the mean of predicted 21-day repairs jobs minus two standard deviations of the mean.  The upper is set to equal to the mean of predicted 21-day repairs jobs plus two standard deviations of the mean.   The current criterion to overwrite the out of limits values is set to replace the value below the lower limit with the lowest value predicted in the corresponding 21-day period that is within the limits, and to replace the value above the upper limit with the highest value predicted in the 21-day period that is within the limits.

##A Note on Model Production Deployment
The model developed in this project is applicable to a specific geographical area.  In general, there will be a forecasting model needed for each different geographical area for a nationwide implementation because, for example, input data for the model changes per geographical area and different networks could have different characteristics in different areas, especially with regards to weather.   Therefore, it is expected that several hundreds of these models will need to be implemented at the same time in a production environment for a nationwide deployment.  The process of developing each model is the same as the one used in this project and the difference in the models could be different beta sets (coefficients) and corresponding input variables, when applicable.  The production deployment plan for a nationwide rollout would be to separate the build and train tasks of the models in an “analytical” environment from the models scoring which would be in a production-standard environment.  One of the business requirements of this project is to be able to productize the model(s) in a standard relational database engine such as ORACLE or Teradata.  The linear regression approach used in our development efforts supports this objective because the beta sets can be developed separately and shipped to a database where the extraction, transformation, and loading (ETL) process for the model(s) input takes place.  The model(s) scoring also takes place in the same production environment where the terms of the betas and corresponding ETL input variables are multiplied and added to generate the model(s) predictions.

```python
#for history weather info use this URL
#http://api.openweathermap.org/data/2.5/forecast/daily?lat=42.25&lon=-71&cnt=16&mode=json

#read the weather info from the web (up to 11 days forecast)
url_str = "http://api.openweathermap.org/data/2.5/forecast/daily?lat=42.25&lon=-71&cnt=11&mode=json"
info = requests.get(url_str).json()
```

```python
#extrat only the weather forecast info
wforecast = info['list']

#count the number of forecast days (up to 11 days)
forecast_days = len(wforecast)

#initialize a dictionary to extract only the rain information
rain_forecast = {}

#loop thru for 21 days
for i in range(21): 
    
    #set the rain info based on forecast weather
    if i < forecast_days: 
        
        #convert the forecast date to local timezone
        local_fdt = pytz.utc.localize(datetime.datetime.utcfromtimestamp(wforecast[i]['dt'])).astimezone(get_localzone())
        
        #create dictionary of binary rain information for 21 days 
        rain_forecast[local_fdt.date()] = 1 if 'rain' in wforecast[i].keys() else 0
             
#convert the dict to data frame
dfw = pd.DataFrame.from_dict((list(rain_forecast.iteritems())))

#rename the columns
dfw.columns=['dt','wx_rain']

#calculate the day of week
dfw['dt'] =  map(lambda x: pd.to_datetime(x), dfw.dt)
dfw['dayofweek'] = map(lambda x: x.dayofweek, dfw.dt)
dfw['weekday'] = map(lambda x: weekday[x.dayofweek], dfw.dt)

#set the dt as index
dfw.set_index('dt',inplace = True)

#sort the 21 days forecast
dfw.sort(inplace=True)

#show the head of the dataframe
dfw.head()
```

```python
#get the max forecast date
max_fdt = dfw.index.max()

#add the remaining days and set the rain to 0
for i in range(21 - dfw.shape[0]):
    
    #set the date
    fdt = max_fdt + np.timedelta64(i + 1, 'D')
    
    #get the day of week
    fdow = fdt.dayofweek
    
    #append to dataframe
    dfw.loc[fdt] = [0, fdow, weekday[fdow]]

#add isholiday column to the dataframe    
dfw['isholiday'] = map(lambda x: 1 if x in us_holidays else 0, dfw.index.date)

#get the holiday plus 1 by shift is holiday
dfw['ishol_plus1'] = dfw.isholiday.shift(1)

#fill the nan after shift
dfw.loc[dfw.index.min(), 'ishol_plus1'] = 1 if dfw.index.min() in us_holidays else 0

#convert the weekday categorical into dummy variables
dfw = pd.get_dummies(dfw, prefix=['day'], columns=['weekday'])

#rename column names to lower case
dfw.columns = [x.lower() for x in dfw.columns]

#show the head of the dataframe
dfw.head()
```
